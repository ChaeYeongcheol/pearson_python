```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import VotingRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score

# 데이터 로드
data = pd.read_csv('result__20231024_VI02.csv')

# 'Loading Rate'와 'Retest Rate'와 상관관계가 높은 상위 특성 선택
selected_features = [
    "'Linear", "'Bowl", "'Electrification",
    't_1_gr_ul', 't_2_tr_ul', 't_1_tr_ul', 'l_1_gr_max', 'thickness',
    't_1_gr_avr', 'l_1_gr_ul', 't_1_tr_avr', 't_1_tr_max', 'l_1_tr_max',
    'gr_cap_min_100', 'gr_min', 'tr_cap_min_100', 'gr_avg', 'cast_dsgn_thk_type',
    'cast_dsgn_thk', 'gr_max', 'total_top_cover', 'top_cover_thk', 'bot_cover_thk',
]

# 특성과 타겟 변수 선택
features = data[selected_features]
target = data[['\'LoadingRate', '\'RetestRate']]

# 범주형 변수와 수치형 변수 분리
categorical_features = features.select_dtypes(include=['object']).columns
numeric_features = features.select_dtypes(include=['int64', 'float64']).columns

# 데이터를 훈련 세트와 테스트 세트로 분할
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# 전처리: 범주형 변수 One-hot 인코딩
preprocessor = ColumnTransformer(
    transformers=[
        ('num', SimpleImputer(strategy='mean'), numeric_features),
        ('cat', OneHotEncoder(), categorical_features)
    ])

# 훈련 데이터와 테스트 데이터를 합쳐서 전처리
X_all = pd.concat([X_train, X_test])
X_all_processed = preprocessor.fit_transform(X_all)

# MLPRegressor 모델 초기화
mlp_regressor = MLPRegressor(max_iter=10000)

# 탐색할 하이퍼파라미터 그리드 설정
param_grid = {
    'hidden_layer_sizes': [(50,), (100,), (50, 50), (100, 100)],
    'alpha': [0.0001, 0.001, 0.01],
    'learning_rate_init': [0.001, 0.01, 0.1]
}

# 그리드 서치 정의
grid_search = GridSearchCV(mlp_regressor, param_grid, cv=5, scoring='neg_mean_squared_error', error_score='raise')

# 그리드 서치 실행
grid_search.fit(X_all_processed[:len(X_train)], y_train['\'LoadingRate'])

# 최적의 하이퍼파라미터 출력
print("loading rate 최적 파라미터:", grid_search.best_params_)

# 최적의 하이퍼파라미터를 사용하여 MLPRegressor 초기화
best_mlp_regressor = MLPRegressor(max_iter=10000, **grid_search.best_params_)

# Voting Regressor 모델 초기화
voting_regressor_loading_rate = VotingRegressor([('gb', GradientBoostingRegressor()), ('mlp', best_mlp_regressor)])


# 그리드 서치 실행
grid_search.fit(X_all_processed[:len(X_train)], y_train['\'RetestRate'])

# 최적의 하이퍼파라미터 출력
print("retest rate 최적 파라미터:", grid_search.best_params_)

# 최적의 하이퍼파라미터를 사용하여 MLPRegressor 초기화
best_mlp_regressor = MLPRegressor(max_iter=10000, **grid_search.best_params_)

voting_regressor_retest_rate = VotingRegressor([('gb', GradientBoostingRegressor()), ('mlp', best_mlp_regressor)])

# 훈련 데이터에 대해 훈련된 Voting Regressor 모델 훈련
voting_regressor_loading_rate.fit(X_all_processed[:len(X_train)], y_train['\'LoadingRate'])
voting_regressor_retest_rate.fit(X_all_processed[:len(X_train)], y_train['\'RetestRate'])

# 테스트 데이터에 대해 평가
y_pred_loading_rate = voting_regressor_loading_rate.predict(X_all_processed[len(X_train):])
y_pred_retest_rate = voting_regressor_retest_rate.predict(X_all_processed[len(X_train):])

mse_loading_rate = mean_squared_error(y_test['\'LoadingRate'], y_pred_loading_rate)
mse_retest_rate = mean_squared_error(y_test['\'RetestRate'], y_pred_retest_rate)

r2_loading_rate = r2_score(y_test['\'LoadingRate'], y_pred_loading_rate)
r2_retest_rate = r2_score(y_test['\'RetestRate'], y_pred_retest_rate)

print("Voting Regressor 'LoadingRate' MSE:", mse_loading_rate)
print("Voting Regressor 'LoadingRate' R-squared:", r2_loading_rate)
print("Voting Regressor 'RetestRate' MSE:", mse_retest_rate)
print("Voting Regressor 'RetestRate' R-squared:", r2_retest_rate)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 손실 그래프 그리기
plt.figure(figsize=(12, 5))

# Gradient Boosting Regressor의 손실 그래프
plt.subplot(2, 2, 1)
for estimator in voting_regressor_loading_rate.estimators_:
    if isinstance(estimator, GradientBoostingRegressor):
        plt.plot(np.arange(1, estimator.n_estimators + 1),
                 estimator.train_score_,
                 label='Training Loss')
plt.xlabel('Number of Estimators')
plt.ylabel('Training Loss')
plt.title('Training Loss Curve - Gradient Boosting Regressor (LoadingRate)')
plt.legend()

# Gradient Boosting Regressor의 손실 그래프
plt.subplot(2, 2, 2)
for estimator in voting_regressor_retest_rate.estimators_:
    if isinstance(estimator, GradientBoostingRegressor):
        plt.plot(np.arange(1, estimator.n_estimators + 1),
                 estimator.train_score_,
                 label='Training Loss')
plt.xlabel('Number of Estimators')
plt.ylabel('Training Loss')
plt.title('Training Loss Curve - Gradient Boosting Regressor (RetestRate)')
plt.legend()


# MLP Regressor의 손실 그래프
plt.subplot(2, 2, 3)
# 반복 횟수에 따른 손실 정보 기록
mlp_regressor_loss = []
for i in range(best_mlp_regressor.max_iter):
    best_mlp_regressor.partial_fit(X_all_processed[:len(X_train)], y_train['\'LoadingRate'])
    mlp_regressor_loss.append(best_mlp_regressor.loss_)

plt.plot(np.arange(1, best_mlp_regressor.max_iter + 1), mlp_regressor_loss, label='Training Loss')
plt.xlabel('Number of Iterations')
plt.ylabel('Training Loss')
plt.title('Training Loss Curve - MLP Regressor (LoadingRate)')
plt.legend()

# MLP Regressor의 손실 그래프
plt.subplot(2, 2, 4)
# 반복 횟수에 따른 손실 정보 기록
mlp_regressor_loss = []
for i in range(best_mlp_regressor.max_iter):
    best_mlp_regressor.partial_fit(X_all_processed[:len(X_train)], y_train['\'RetestRate'])
    mlp_regressor_loss.append(best_mlp_regressor.loss_)

plt.plot(np.arange(1, best_mlp_regressor.max_iter + 1), mlp_regressor_loss, label='Training Loss')
plt.xlabel('Number of Iterations')
plt.ylabel('Training Loss')
plt.title('Training Loss Curve - MLP Regressor (RetestRate)')
plt.legend()

plt.tight_layout()
plt.show()

# X_train의 첫 번째 샘플 추출
first_sample = X_test.iloc[[3]]

# 모델에 첫 번째 샘플을 넣고 예측 수행
loading_rate_prediction = voting_regressor_loading_rate.predict(preprocessor.transform(first_sample))
retest_rate_prediction = voting_regressor_retest_rate.predict(preprocessor.transform(first_sample))

# 예측 결과 출력
print("원본 : ", y_test.iloc[3])
print("예측 'LoadingRate':", loading_rate_prediction[0])
print("예측 'RetestRate':", retest_rate_prediction[0])
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import VotingRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score

# 데이터 로드
data = pd.read_csv('result__20231024_VI01.csv')

# 'Loading Rate'와 'Retest Rate'와 상관관계가 높은 상위 특성 선택
selected_features_loading = [
    "'Linear", "'Bowl", "'Electrification",
    't_1_gr_ul', 't_2_tr_ul', 't_1_tr_ul', 'l_1_gr_max', 'thickness',
    't_1_gr_avr', 'l_1_gr_ul', 't_1_tr_avr', 't_1_tr_max', 'l_1_tr_max'
]
selected_features_retest = [
    "'Linear", "'Bowl", "'Electrification",
    'gr_cap_min_100', 'gr_min', 'tr_cap_min_100', 'gr_avg', 'cast_dsgn_thk_type',
    'cast_dsgn_thk', 'gr_max', 'total_top_cover', 'top_cover_thk', 'bot_cover_thk'
]

# 특성과 타겟 변수 선택
features_loading = data[selected_features_loading]
target_loading = data[['\'LoadingRate']]

features_retest = data[selected_features_retest]
target_retest = data[['\'RetestRate']]

# 범주형 변수와 수치형 변수 분리
categorical_features_loading = features_loading.select_dtypes(include=['object']).columns
numeric_features_loading = features_loading.select_dtypes(include=['int64', 'float64']).columns
categorical_features_retest = features_retest.select_dtypes(include=['object']).columns
numeric_features_retest = features_retest.select_dtypes(include=['int64', 'float64']).columns

# 데이터를 훈련 세트와 테스트 세트로 분할
X_train_loading, X_test_loading, y_train_loading, y_test_loading = train_test_split(features_loading, target_loading, test_size=0.2, random_state=42)

# 데이터를 훈련 세트와 테스트 세트로 분할
X_train_retest, X_test_retest, y_train_retest, y_test_retest = train_test_split(features_retest, target_retest, test_size=0.2, random_state=42)

# 전처리: 범주형 변수 One-hot 인코딩
preprocessor_loading = ColumnTransformer(
    transformers=[
        ('num', SimpleImputer(strategy='mean'), numeric_features_loading),
        ('cat', OneHotEncoder(), categorical_features_loading)
    ])
# 전처리: 범주형 변수 One-hot 인코딩
preprocessor_retest = ColumnTransformer(
    transformers=[
        ('num', SimpleImputer(strategy='mean'), numeric_features_retest),
        ('cat', OneHotEncoder(), categorical_features_retest)
    ])

# 훈련 데이터와 테스트 데이터를 합쳐서 전처리
X_all_loading = pd.concat([X_train_loading, X_test_loading])
X_all_processed_loading = preprocessor_loading.fit_transform(X_all_loading)

# 훈련 데이터와 테스트 데이터를 합쳐서 전처리
X_all_retest = pd.concat([X_train_retest, X_test_retest])
X_all_processed_retest = preprocessor_retest.fit_transform(X_all_retest)

# MLPRegressor 모델 초기화
mlp_regressor = MLPRegressor(max_iter=10000)

# 탐색할 하이퍼파라미터 그리드 설정
param_grid = {
    'hidden_layer_sizes': [(50,), (100,), (50, 50), (100, 100)],
    'alpha': [0.0001, 0.001, 0.01],
    'learning_rate_init': [0.001, 0.01, 0.1]
}

# 그리드 서치 정의
grid_search = GridSearchCV(mlp_regressor, param_grid, cv=5, scoring='neg_mean_squared_error', error_score='raise')

# 그리드 서치 실행
grid_search.fit(X_all_processed_loading[:len(X_train_loading)], y_train_loading)

# 최적의 하이퍼파라미터 출력
print("loading rate 최적 파라미터:", grid_search.best_params_)

# 최적의 하이퍼파라미터를 사용하여 MLPRegressor 초기화
best_mlp_regressor = MLPRegressor(max_iter=10000, **grid_search.best_params_)

# Voting Regressor 모델 초기화
voting_regressor_loading_rate = VotingRegressor([('gb', GradientBoostingRegressor()), ('mlp', best_mlp_regressor)])


# 그리드 서치 실행
grid_search.fit(X_all_processed_retest[:len(X_train_retest)], y_train_retest)

# 최적의 하이퍼파라미터 출력
print("retest rate 최적 파라미터:", grid_search.best_params_)

# 최적의 하이퍼파라미터를 사용하여 MLPRegressor 초기화
best_mlp_regressor = MLPRegressor(max_iter=10000, **grid_search.best_params_)

voting_regressor_retest_rate = VotingRegressor([('gb', GradientBoostingRegressor()), ('mlp', best_mlp_regressor)])

# 훈련 데이터에 대해 훈련된 Voting Regressor 모델 훈련
voting_regressor_loading_rate.fit(X_all_processed_loading[:len(X_train_loading)], y_train_loading)
voting_regressor_retest_rate.fit(X_all_processed_retest[:len(X_train_retest)], y_train_retest)

# 테스트 데이터에 대해 평가
y_pred_loading_rate = voting_regressor_loading_rate.predict(X_all_processed_loading[len(X_train_loading):])
y_pred_retest_rate = voting_regressor_retest_rate.predict(X_all_processed_retest[len(X_train_retest):])

mse_loading_rate = mean_squared_error(y_test_loading, y_pred_loading_rate)
mse_retest_rate = mean_squared_error(y_test_retest, y_pred_retest_rate)

r2_loading_rate = r2_score(y_test_loading, y_pred_loading_rate)
r2_retest_rate = r2_score(y_test_retest, y_pred_retest_rate)

print("Voting Regressor 'LoadingRate' MSE:", mse_loading_rate)
print("Voting Regressor 'LoadingRate' R-squared:", r2_loading_rate)
print("Voting Regressor 'RetestRate' MSE:", mse_retest_rate)
print("Voting Regressor 'RetestRate' R-squared:", r2_retest_rate)
```
